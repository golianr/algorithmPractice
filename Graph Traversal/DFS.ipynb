```python
# Graph init.

import networkx as nx
import matplotlib.pyplot as plt

graph = nx.Graph()

graph.add_nodes_from([0, 1, 2, 3, 4, 5, 6, 7])

graph.add_edges_from([
    (0, 1),
    (0, 2),
    (1, 3),
    (1, 4),
    (2, 5),
    (2, 6),
    (3, 7)
])

a = nx.draw(graph, with_labels=True)

plt.show(a)
```

```python
# DFS
def dfs(graph, node, visited, path):
    
    if visited[node] is True:
        return
    visited[node] = True
    path.append(node)
    
    for neighbor in graph.neighbors(node):
        dfs(graph, neighbor, visited, path)

    return path
```

```python
# Init. of DFS
path = dfs(graph, 0, [False] * 8, [])
print(f"Order of DFS from 0: {path}")
```

```python
# BFS
def bfs(graph, start, visited, path):
    stack = [start]
    while stack:
        current = stack.pop(0)
        path.append(current)
        visited[current] = True

        for neighbor in graph.neighbors(current):
            if visited[neighbor] is False:
                stack.append(neighbor)

    return path
```

```python
# Init. of BFS
path = bfs(graph, 0, [False] * 8, [])
print(path)
```
